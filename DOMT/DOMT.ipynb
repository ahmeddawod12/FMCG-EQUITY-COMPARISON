# المقدمه

تأسّست في 14 مارس 1984 على يد عمر محمد عبد الحميد الدمياطي، ومقرّها الجيزة. وهي أكبر منتج للجبن في مصر من حيث الحصة السوقية، وأُدرجت في البورصة في 2016 بطرح 49% من أسهمها بقيمة تتراوح بين 1.1 و1.2 مليار جنيه.  (Economies)
تعمل دومتي كمنتج ألبان متكامل عبر ثلاثة قطاعات — الألبان والعصائر والمخبوزات — وتقدّم الجبن الأبيض والمطبوخ والموتزاريلا والعصائر ومنتجات المخبوزات (سندوتشات وخبز وكرواسون) واللبن، تحت علامات دومتي وبرافو وسليم، وتصدّر لنحو 35 دولة.  (TradingView) وتُعدّ منذ 2012 أكبر منتج في العالم للجبن الأبيض المعبّأ في تيتراباك، وتسيطر على أكثر من 35% من سوق الجبن الأبيض غير المطبوخ في مصر.  (Economies)
موقعها في التحليل — دومتي هي الشركة الثالثة في المقارنة (مع إيديتا وجهينة)، والسؤال واحد: أيّها أقدر على التحكّم في ربحه وأصمد في الأزمات؟ ومن النظرة الأولى تبرز ثلاث ملاحظات: نموٌّ تضاعف 2.8 مرة ثم توقّف في 2025 (+1.8%)؛ صافي ربح انهار 68% رغم تحسّن الهامش الإجمالي؛ وعبء تمويلي التهم 69% من الربح التشغيلي وأنزل تغطية الفوائد إلى 1.46x — أضعف مركز مالي بين الثلاثة.
